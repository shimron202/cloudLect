```python
import json
import ipywidgets as widgets
from IPython.display import display
import random
from IPython.display import Javascript



class Question:
    def __init__(self, description="", options=None, correct_answer_index=-1):
        self.description = description
        self.options = options or []
        self.correct_answer_index = correct_answer_index
file_path = '/content/drive/My Drive/questions.json'
my_questions = [] # empty array of type Question
currrent_question_number=0
#elements---
question_label = widgets.Label(value="Question ")
question_number = widgets.Label(value="0")
question = widgets.Textarea(
    value="",
    description="",
    layout={'height': 'auto', 'width': 'auto'},  # Set height to auto for automatic expansion
)
options = widgets.RadioButtons(options=[], description="",tooltips=["A ","B ","C ","D " ])
answer = widgets.Button(description="answer")
points = widgets.Label(value="0",description = "points  ")
label= widgets.Label(value="points ")


hbox = widgets.HBox([question_label,question_number])
widgets_vbox = widgets.HBox([answer,label,points])
widgets_box = widgets.VBox([hbox,question,options,widgets_vbox])

# ----

#style
answer.layout.margin='0 80px 0 0'
hbox.layout.margin = '0px 0 10px 0'
question.layout.margin = '0 0 20px 0'
options.layout.margin = ' 0 0 30px 0 '
#----

#functions---
def getQuestionFromDrive():
  global my_questions
  questions_json=[] # save the data of question in json foramt
  with open(file_path, 'r') as question_file:  # update file students.json
      questions_json=json.load(question_file) # load get  file handler and insert all the json data to the array
  for question in questions_json.get("question_info", []):
    new_question = Question(description=question["description"],options=question["options"],correct_answer_index=int(question["correct_answer"])) # create instance of Quetion
    my_questions.append(new_question) # add that new question to the array

def setData():
  global currrent_question_number
  question_to_display = my_questions[currrent_question_number]
  currrent_question_number+=1
  question_number.value=str(currrent_question_number)
  question.value = question_to_display.description
  options.options=question_to_display.options

def check_answer(): # reutrn true if the answer is correct else false
  index_of_choosed_answer = options.index
  return index_of_choosed_answer == my_questions[currrent_question_number-1].correct_answer_index

def on_answer_clicked(b):
  correct_answer = options.options[my_questions[currrent_question_number-1].correct_answer_index]

  if(check_answer()):
     display(Javascript('alert("you are correct!");'))
     points.value=str(int(points.value)+20)
  else:
      display(Javascript(f'alert("Wrong. The correct answer is: {correct_answer}");'))
  if(currrent_question_number<len(my_questions)):
    setData()
  else:
    display(Javascript(f'alert("Total points: {points.value}");'))
    reset()

def reset():
  global my_questions
  global currrent_question_number
  random.shuffle(my_questions)
  currrent_question_number=0
  points.value="0"
  setData()


# ----
getQuestionFromDrive()
reset()
answer.on_click(on_answer_clicked)

# Create an output widget to display the selected option
output = widgets.Output()

display(widgets_box,output)
```
